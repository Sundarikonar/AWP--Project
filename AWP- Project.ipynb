```python
#Saving the mean of total revenue in mean variable
revenue_mean=df['Total Revenue'].mean()
profit_mean=df['Total Profit'].mean()
```

```python
plt.figure(figsize=(5,5))
explode=[0.05,0,0,0,0,0,0]
df['Region'].value_counts().plot(kind='pie',autopct='%.2f%%',explode=explode)
plt.title('Region wise Sales')
plt.show()
```

From the above pie chart we can observe that the maximum sales is from Europe region followed by Africa 
NorthAmerica has the lowest sales count

```python
plt.figure(figsize=(25,7))
sns.countplot(df['Item Type'])
plt.title('Item type wise Sales count')
plt.show()
```

We can observe above that the maximum sales is of Beverages followed by Vegetables, Office supplies
Lowest item type sale is of Fruits and cosmetics

```python
plt.figure(figsize=(8,8))
sns.countplot(data=df,y=df['Region'],hue='Item Type')
plt.show()
```

In the above detailed analysis we can observe that in Europe region people buy vegetables the most
In African region people consume fruits the most
Asian region sell most of beverages and cosmetics
Middle east and North African region sell most of beverages, vegetables and clothes
Central American region sell household supplies the most
Australian Region sells balances amount of Vegetables, household, Snacks and personal care
whereas North American region sells Clothes and baby food the most, but none of cosmetics and cereal

```python
#Same above visualisation is presented below in different format
plt.figure(figsize=(30,5))
sns.countplot(data=df,x=df['Item Type'],hue='Region')
plt.show()
```

```python
plt.figure(1,figsize=(15,15))

plt.subplot(3,2,1)
plt.title('Unit Cost Analysis')
sns.lineplot(data=df,x='Year',y='Unit Cost',hue='Sales Channel')

plt.subplot(3,2,2)
plt.title('Total Cost Analysis')
sns.lineplot(data=df,x='Year',y='Total Cost',hue='Sales Channel')

plt.subplot(3,2,3)
plt.title('Unit Price Analysis')
sns.lineplot(data=df,x='Year',y='Unit Price',hue='Sales Channel')

plt.subplot(3,2,4)
plt.title('Total Revenue Analysis')
sns.lineplot(data=df,x='Year',y='Total Revenue',hue='Sales Channel')

plt.subplot(3,2,5)
plt.title('Unit Sales Analysis')
sns.lineplot(data=df,x='Year',y='Units Sold',hue='Sales Channel')

plt.subplot(3,2,6)
plt.title('Profit Analysis')
sns.lineplot(data=df,x='Year',y='Total Profit',hue='Sales Channel')

plt.show()
```
